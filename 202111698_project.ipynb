```python
# 3번
import pandas as pd

data = {
    '학번': [202303, 202405, 202407, 202409],
    '이름': ['오지환', '박용택', '정성훈', '다니엘'],
    '트랙': ['데이터', '데이터', '데이터', '생산'],
    'Quiz': [15, 20, 10, 17],
    '중간고사': [20, 20, 15, 18]
}

df = pd.DataFrame(data)

from IPython.display import display
display(df)
```

```python
# 4번
new_row = pd.DataFrame([{'학번': 202502, '이름': '지민', '트랙': '생산', 'Quiz': 18, '중간고사': 16}])

df = pd.concat([df, new_row], ignore_index=True)

from IPython.display import display
display(df)
```

```python
#5번
data = {
    '학번': [202303, 202405, 202407, 202409, 202502],
    '이름': ['오지환', '박용택', '정성훈', '다니엘', '지민'],
    '트랙': ['데이터', '데이터', '데이터', '생산', '생산'],
    'Quiz': [15, 20, 10, 17, 18],
    '중간고사': [20, 20, 15, 18, 16]
}

df = pd.DataFrame(data)

# '정성훈' 학생의 데이터를 제거
df2 = df[df['이름'] != '정성훈'].reset_index(drop=True)

display(df2)
```

```python
#6번
result = df2[['이름', '중간고사']]

print(result)
```

```python
#7번
df = df.reindex([2, 0, 4, 1, 3])

display(df)
```

```python
#8번
from IPython.display import display

record_iloc = df.iloc[3]

record_loc = df.loc[1]

display("Record using iloc:", record_iloc)
display("Record using loc:", record_loc)
```

```python
#9번
result_iloc = df.iloc[:, [1, 4]]

result_loc = df.loc[:, ['이름', '중간고사']]

from IPython.display import display
display("Result using iloc:", result_iloc)
display("Result using loc:", result_loc)
```

```python
#10번
midterm_mean = df['중간고사'].mean()

midterm_mean
```

```python
# 11번
lowest_midterm_index = df['중간고사'].idxmin()

# 결과 출력
lowest_midterm_index
```

```python
# 12번
high_student_id = df.loc[df['Quiz'].idxmax(), '학번']

high_student_id
```

```python
# 13번
statistics_summary = df.describe()

display(statistics_summary)
```

```python
# 14번
df_sorted = df.sort_values(by='중간고사', ascending=False)

from IPython.display import display
display(df_sorted)
```

```python
#15 번
quiz_mean = df['Quiz'].mean()

students_above_average = df[df['Quiz'] >= quiz_mean]['이름']

from IPython.display import display
display(students_above_average)
```
